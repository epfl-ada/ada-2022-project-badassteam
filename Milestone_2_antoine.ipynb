```python
def stringfreebase(x):
    x = x[27:]
    x = x.replace('.','/')
    x = x.replace('>','')
    return x
```

```python
def getid(x):
    try:
        client = Client()
        entity = client.get(x, load=True)
        freebase_id = client.get('P646')
        freebase_id = entity[freebase_id]
        try:
            review = client.get('P444')
            review = entity[review]
        except:
            review = 0
        ind = film_us[film_us.item==x].index[0]
        if ind % 10 == 0:
           print(ind)
        return pd.Series([freebase_id,review])
    except:
        x=0
        return pd.Series([x,x])
```

```python
film_us[['freebase_ID','review']] = film_us.item.apply(getid)
```
