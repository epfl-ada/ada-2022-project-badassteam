```python
film_query = """
SELECT ?item ?itemLabel
WHERE {
SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE]". }
{
SELECT DISTINCT ?item WHERE {
?item p:P444 ?statement0.
?statement0 (ps:P444) _:anyValueP444.
?item p:P31 ?statement1.
?statement1 (ps:P31/(wdt:P279*)) wd:Q11424.
}
}
}
"""
```

```python
film_df = wikidata2df(film_query)
```

```python
film_df
```

```python
mergedall = movies.merge(film_df,left_on='wiki',right_on='item')
mergedall = mergedall.drop(columns=['item','itemLabel','freebase'])
mergedall
```

```python
def getdata(x):
    client = Client()
    entity = client.get(x, load=True)
    spouse_prop = client.get('P444')
    spouse_prop = entity[spouse_prop]
    return spouse_prop
```

```python
mergedtest = mergedall.sample(10)
mergedtest['score'] = mergedtest.wiki.apply(getdata)
mergedtest
```

```python
def getid(x):
    try:
        client = Client()
        entity = client.get(x, load=True)
        freebase_id = client.get('P646')
        freebase_id = entity[freebase_id]
        try:
            review = client.get('P444')
            review = entity[review]
        except:
            review = 0
        ind = film_us[film_us.item==x].index[0]
        if ind % 10 == 0:
           print(ind)
        return pd.Series([freebase_id,review])
    except:
        x=0
        return pd.Series([x,x])
```
